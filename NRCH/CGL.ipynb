```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
l1 = lambda r, u, a0, a1, x :\
    r - 2*u*x**2 - np.sqrt((u**2+a1**2)*x**4 - (a0 - 2*a1*x**2)**2 + 0j)
l2 = lambda r, u, a0, a1, x :\
    r - 2*u*x**2 + np.sqrt((u**2+a1**2)*x**4 - (a0 - 2*a1*x**2)**2 + 0j)
```

```python
m = 4
a = np.linspace(.5, 5, m)
a, b = np.meshgrid(a, a)
a = a.flatten()
b = b.flatten()
a = list(zip(a, b))
```

```python
r = u = 1

fig, ax = plt.subplots(m, m, figsize=(30, 30))
for i, (a0, a1) in enumerate(a):
    l = lambda x: l1(r, u, a0, a1, x)
    ll = lambda x: l2(r, u, a0, a1, x)

    x = np.linspace(.2, 2, 500)
    n = (i%m, i//m)
    ax[n].plot(x, ll(x).real, 'k', label='Re $\\lambda_+$')
    ax[n].plot(x, l(x).real, 'r--', label='Re $\\lambda_-$')
    ax[n].plot(x, l(x).imag, 'green', label='Im $\\lambda_+$')
    title ="$\\alpha_0 = {:.2f}, \\alpha_1 = {:.2f}$".format(a0, a1)
    ax[n].set_title(title)
plt.legend()
plt.show()
```
