```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
q = .1

l1 = lambda r, a : r + q + np.sqrt(r**2 - a**2 + 0j)
l2 = lambda r, a : r + q - np.sqrt(r**2 - a**2 + 0j)
```

```python
n = 1000
a = np.linspace(0, 1, n)

for r in [.1, .3, .7, 1.]:
    fig, ax = plt.subplots()
    ax.plot(a, l1(r, a).real, 'k')
    ax.plot(a, l2(r, a).real, 'r')
    ax.plot(a, l1(r, a).imag, 'k--')
    ax.plot(a, l2(r, a).imag, 'r--')
    plt.show()
```
