```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
u = 1.
r = 1.

def f(phi, a):
    p2 = phi[0]**2 + phi[1]**2
    mu1 = -r*phi[0] + u * p2 * phi[0] + a*phi[1]
    mu2 = -r*phi[1] + u * p2 * phi[1] - a*phi[0]
    return np.array([mu1, mu2])

```

```python
n = 500
k = 1.2

x1, x2 = np.linspace(-k, k, n), np.linspace(-k, k, n)
x1, x2 = np.meshgrid(x1, x2)

aa = np.linspace(0, 1.2, 9)
fig, ax = plt.subplots(3, 3, figsize=(30, 30))
for n, a in enumerate(aa):
    v1, v2 = - f([x1, x2], a)
    l = np.sqrt(v1**2 + v2**2)
    ax[n//3, n%3].streamplot(x1, x2, v1, v2, color=l, density=2., linewidth=1)
    ax[n//3, n%3].set_title("$\\alpha={:.2f}$".format(a))
# plt.tight_layout()
```
